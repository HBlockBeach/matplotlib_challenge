## Observations and Insights

Observation #1: Ramicane provides the most consistent tumor volume reduction of the treatment regimens. It has the closest mean and median, as well as the lowest variance and the smallest standard deviation.

Observation #2: Mouse S185 had tumor volume reduction at each timepoint other than timpoint 15. 

Observation #3: For mice treated with Capomuilin, there is a strong correlation between average tumor volume and weight. 

Would be interesting to see if there was an even distribution of male and female mice in each group. If so, could also say that gender has no effect on the drugs effectiveness.

```python
# Calculate the final tumor volume of each mouse across four of the most promising treatment regimens. Calculate the IQR and quantitatively determine if there are any potential outliers. 

#ramicane
ramicane_final_volume = ramicane[["Mouse_ID", "Timepoint", "TV"]]

ramicane_group = ramicane_final_volume.groupby("Mouse_ID")
max_timepoint_ramicane = ramicane_group["Timepoint"].max()
max_timepoint_ramicane = pd.DataFrame(max_timepoint_ramicane)

merged_max_ramicane = pd.merge(max_timepoint_ramicane, ramicane, on = ("Mouse_ID","Timepoint"), how = "left")
final_tumor_volume_ramicane = merged_max_ramicane[["Mouse_ID","TV"]]


#capomulin
capomulin_final_volume = capomulin[["Mouse_ID", "Timepoint", "TV"]]

capomulin_group = capomulin_final_volume.groupby("Mouse_ID")
max_timepoint_capomulin = capomulin_group["Timepoint"].max()
max_timepoint_capomulin = pd.DataFrame(max_timepoint_capomulin)

merged_max_capomulin = pd.merge(max_timepoint_capomulin, capomulin, on = ("Mouse_ID","Timepoint"), how = "left")
final_tumor_volume_capomulin = merged_max_capomulin[["Mouse_ID","TV"]]


#infubinol
infubinol_final_volume = infubinol[["Mouse_ID", "Timepoint", "TV"]]

infubinol_group = infubinol_final_volume.groupby("Mouse_ID")
max_timepoint_infubinol = infubinol_group["Timepoint"].max()
max_timepoint_infubinol = pd.DataFrame(max_timepoint_infubinol)

merged_max_infubinol = pd.merge(max_timepoint_infubinol, infubinol, on = ("Mouse_ID","Timepoint"), how = "left")
final_tumor_volume_infubinol = merged_max_infubinol[["Mouse_ID","TV"]]


#ceftamin
ceftamin_final_volume = ceftamin[["Mouse_ID", "Timepoint", "TV"]]

ceftamin_group = ceftamin_final_volume.groupby("Mouse_ID")
max_timepoint_ceftamin = ceftamin_group["Timepoint"].max()
max_timepoint_ceftamin = pd.DataFrame(max_timepoint_ceftamin)

merged_max_ceftamin = pd.merge(max_timepoint_ceftamin, ceftamin, on = ("Mouse_ID","Timepoint"), how = "left")
final_tumor_volume_ceftamin = merged_max_ceftamin[["Mouse_ID","TV"]]


#tables
#final_tumor_volume_ramicane
#final_tumor_volume_capomulin
#final_tumor_volume_infubinol
#final_tumor_volume_ceftamin




#ramicane
quartiles_ramicane = final_tumor_volume_ramicane["TV"].quantile([.25,.5,.75])
lowerq_ramicane = quartiles_ramicane[0.25]
upperq_ramicane = quartiles_ramicane[0.75]
iqr_ramicane = upperq_ramicane-lowerq_ramicane
lower_bound_ramicane = lowerq_ramicane - (1.5*iqr_ramicane)
upper_bound_ramicane = upperq_ramicane + (1.5*iqr_ramicane)

ramicane_counter = 0
print("------------Ramicane------------")
for x in final_tumor_volume_ramicane["TV"]:
    if x < lower_bound_ramicane:
        print(f"{round(x,2)} is a lower outlier for Ramicane")
        ramicane_counter= ramicane_counter+1
    elif x > upper_bound_ramicane:
        print(f"{round(x,2)} is an upper outlier for Ramicane")
        ramicane_counter = ramicane_counter+1
    else:
        print(f"{round(x,2)} is not an outlier for Ramicane")

#capomulin
quartiles_capomulin = final_tumor_volume_capomulin["TV"].quantile([.25,.5,.75])
lowerq_capomulin = quartiles_capomulin[0.25]
upperq_capomulin = quartiles_capomulin[0.75]
iqr_capomulin = upperq_capomulin-lowerq_capomulin
lower_bound_capomulin = lowerq_capomulin - (1.5*iqr_capomulin)
upper_bound_capomulin = upperq_capomulin + (1.5*iqr_capomulin)

capomulin_counter = 0
print("------------Capomulin------------")
for x in final_tumor_volume_capomulin["TV"]:
    if x < lower_bound_capomulin:
        print(f"{round(x,2)} is a lower outlier for Capomulin")
        capomulin_counter=capomulin_counter+1
    elif x > upper_bound_capomulin:
        print(f"{round(x,2)} is an upper outlier for Capomulin")
        capomulin_counter = capomulin_counter+1
    else:
        print(f"{round(x,2)} is not an outlier for Capomulin")
        
#infubinol
quartiles_infubinol = final_tumor_volume_infubinol["TV"].quantile([.25,.5,.75])
lowerq_infubinol = quartiles_infubinol[0.25]
upperq_infubinol = quartiles_infubinol[0.75]
iqr_infubinol = upperq_infubinol-lowerq_infubinol
lower_bound_infubinol = lowerq_infubinol - (1.5*iqr_infubinol)
upper_bound_infubinol = upperq_infubinol + (1.5*iqr_infubinol)

infubinol_counter = 0
print("------------Infubinol------------")
for x in final_tumor_volume_infubinol["TV"]:
    if x < lower_bound_infubinol:
        print(f"{round(x,2)} is a lower outlier for Infubinol")
        infubinol_counter=infubinol_counter+1
    elif x > upper_bound_infubinol:
        print(f"{round(x,2)} is an upper outlier for Infubinol")
        infubinol_counter = infubinol_counter+1
    else:
        print(f"{round(x,2)} is not an outlier for Infubinol")
        
        
#ceftamin
quartiles_ceftamin = final_tumor_volume_ceftamin["TV"].quantile([.25,.5,.75])
lowerq_ceftamin = quartiles_ceftamin[0.25]
upperq_ceftamin = quartiles_ceftamin[0.75]
iqr_ceftamin = upperq_ceftamin-lowerq_ceftamin
lower_bound_ceftamin = lowerq_ceftamin - (1.5*iqr_ceftamin)
upper_bound_ceftamin = upperq_ceftamin + (1.5*iqr_ceftamin)

ceftamin_counter = 0
print("------------Ceftamin------------")
for x in final_tumor_volume_ceftamin["TV"]:
    if x < lower_bound_ceftamin:
        print(f"{round(x,2)} is a lower outlier for Ceftamin")
        ceftamin_counter=ceftamin_counter+1
    elif x > upper_bound_ceftamin:
        print(f"{round(x,2)} is an upper outlier for Ceftamin")
        ceftamin_counter = ceftamin_counter+1
    else:
        print(f"{round(x,2)} is not an outlier for Ceftamin")
        
        
print("------------Totals------------")       
print(f"Ramicane has {ramicane_counter} outliers.")
print(f"Capomulin has {capomulin_counter} outliers.")
print(f"Infubinol has {infubinol_counter} outliers.")
print(f"Ceftamin has {ceftamin_counter} outliers.")
```

```python
# Calculate the correlation coefficient and linear regression model for mouse weight and average tumor volume for the Capomulin regimen

correlation = st.pearsonr(weight["TV_x"],weight["Weight (g)"])
correlation_value = correlation[0]

x_values = weight['Weight (g)']
y_values = weight['TV_x']

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)

regress_values = x_values * slope + intercept

#y=mx+b
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
```
